```
# Importando librerias
import numpy as np # computing scientific with python (arrays, matrices, vectores)
import scipy as sc # computing scientific (imagene, numpy, pandas,... etc) NumPy extendido
import matplotlib.pyplot as plt # graficación
```

Importando data-set de las casas de boston

```
# Importando data-set boston (libreria sklearn) 
# sklearn Machine Learning in Python
from sklearn.datasets import load_boston
```

```
boston = load_boston()
print(boston.DESCR) #descripción de la base
```

**Estimador de mínimos cuadrados**  $\hat{\beta}=(X^{T}X)^{-1} X^{T} Y $

```
X = np.array(boston.data[:, 5]) # número de habitaciones
Y = np.array(boston.target) # valor medio de las viviendas
```

```
# Graficando
plt.scatter(X, Y, alpha = 0.5)
```

```
X = np.array([np.ones(len(X)), X]).T  # transpuesta.
```

```
print(X)
```

```
# Calculando los estiadores de los parametros beta
B = np.linalg.inv(np.transpose(X) @ X) @ np.transpose(X) @ Y
print(B)
```

```
# graficando nuevamente
plt.scatter(X[:, 1], Y, alpha = 0.5)
plt.plot([4, 9], [B[0]+B[1]*4, B[0]+B[1]*9], c= "red")
```
