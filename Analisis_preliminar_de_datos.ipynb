### Borrado de valores faltantes

```
# borrar filas axis=0, columnas axis=1, how="all" (si faltan datos en toda las columnas de esa fila)
data.dropna(axis=0, how="all")
```

```
# Haciendo copia de la base
data2 = data
```

```
# se borra la fila si faltan datos en al menos una de las columnas
data.dropna(axis=0, how="any")
```

## Imputación de datos faltantes
---

```
# Haciendo copia de la base
data3 = data
```

```
# Rellenar dodo con ceros
data3.fillna(0)
```

```
data4 = data
data4["body"] = data4["body"].fillna(0)
data4["home.dest"] = data4["home.dest"].fillna("Desconocido")
data.head()
```

## Imputación por media.

```
pd.isnull(data4["age"]).values.ravel().sum()
```

```
data4["age"] = data4["age"].fillna(data4["age"].mean)
data4.head()
```

## Variables Dummy
---

```
dummy_sex = pd.get_dummies(data4["sex"], prefix="sex")
dummy_sex.head()
```

```
# Eliminar columnas 

data4 = data.drop(["sex"], axis = 1)
```

```
# Concatenar bases 

datos_nuevos = pd.concat([data4, dummy_sex], axis = 1)
datos_nuevos.head(3)
```
