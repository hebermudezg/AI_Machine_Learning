**Par√°metros read_csv()**

```
filepath (Direccion del fichero)
sep (separador)
header (Indice de cabecera)
dtype ("nombre_variable_1":np.float64, "nombre_variable_2":np.int32) # requiere NumPy
names (nombrar variables)
skiprows (saltar filas)
index_col (definir una columna como identificador)
skip_blank_lines=True (la lineas en blanco seran saltadas)
na.filter (Eliminar filas con NaN)
```



````
data = pd.read_csv(filepath_or_buffer="titanic3.csv", sep=",", dtype=None, header=0)
````
